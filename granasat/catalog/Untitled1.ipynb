```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
point_A = [-1, 3, 2]
point_B = [2, 3, 5]
point_C = [3, 5 , -2]
```

```python
x = [point_A[0], point_B[0], point_C[0], point_A[0]]
y = [point_A[1], point_B[1], point_C[1], point_A[1]]
z = [point_A[2], point_B[2], point_C[2], point_A[2]]
```

```python
fig = plt.figure()
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, color = "green")
ax.plot(x,y,z, color='red')
plt.title("3D Triangle")
```

```python
import math

# function for finding the angle  
def angle_triangle(x1, x2, x3, y1, y2, y3, z1, z2, z3):  
  
    num = (x2-x1)*(x3-x1)+(y2-y1)*(y3-y1)+(z2-z1)*(z3-z1)  
  
    den = math.sqrt((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)*\
                math.sqrt((x3-x1)**2+(y3-y1)**2+(z3-z1)**2)  
  
    angle = math.degrees(math.acos(num / den))  
  
    return round(angle, 3)
```

```python
angle_A = angle_triangle(point_A[0], point_B[0], point_C[0], 
                         point_A[1], point_B[1], point_C[2], 
                         point_A[2], point_B[2], point_C[2])
angle_B = angle_triangle(point_B[0], point_C[0], point_A[0], 
                         point_B[1], point_C[1], point_A[1],
                         point_B[2], point_C[2], point_A[2])  
angle_C = angle_triangle(point_C[0], point_B[0], point_A[0],
                         point_C[1], point_B[1], point_A[1],
                         point_C[2], point_B[2], point_A[2])  
print("Angles are :")  
print("angle A = ", angle_A, "degree")  
print("angle B = ", angle_B, "degree")  
print("angle C = ", angle_C, "degree")  
```

```python
# Now lets use Manuels code for finding the angles
# Get vectors between them
import numpy as np
v1 = np.subtract(point_B, point_A)
v2 = np.subtract(point_C, point_A)
v3 = np.subtract(point_C, point_B)
```

```python
# Get angles
angle1 = np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
angle2 = np.dot(-v1, v3)/(np.linalg.norm(v1)*np.linalg.norm(v3))
angle3 = np.dot(-v2, -v3) / (np.linalg.norm(v2) * np.linalg.norm(v3))
```
